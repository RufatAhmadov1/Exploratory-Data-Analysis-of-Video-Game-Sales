```python
data_row_index = games[games['Year'].isnull()].index 
games = games.drop(data_row_index)
data_row_index1 = games[games['Publisher'].isnull()].index 
games = games.drop(data_row_index1)
```

```python
games['Year'] = games['Year'].astype('int64')
```

# How much each genre sold.

```python
sns.set_style('darkgrid')
sns.countplot(y = 'Genre', data = games, order = games['Genre'].value_counts().index, lw =2)
```

# How many games were sold in the top 15 years

```python
games["Year"].value_counts()[:15].plot(kind = "bar", rot = 0, figsize = (15, 8))
plt.title("sales for years" ,fontsize = 19)
plt.xlabel("Years", fontsize = 15, weight = "bold")
plt.ylabel("sales_count", fontsize = 15, weight = "bold");
```

# How many sales were made on each platform.

```python
games["Platform"].value_counts()[:45].plot(kind = "bar", rot = 0, figsize = (15, 8))
plt.title("sales for platforms" ,fontsize = 19)
plt.xlabel("Platform", fontsize = 15, weight = "bold")
plt.ylabel("sales_count", fontsize = 15, weight = "bold");
```

# Sales number of top 10 game distributors

```python
games["Publisher"].value_counts()[:10].plot(kind = "bar", rot = 0, figsize = (15, 8))
plt.title("sales for publishers" ,fontsize = 19)
plt.xlabel("Publisher", fontsize = 15, weight = "bold")
plt.ylabel("sales_count", fontsize = 15, weight = "bold");
```

# How many of the top 5 genres were sold in the top 5 years

```python
top_genres = games['Genre'].value_counts().head(5).index
top_years = games['Year'].value_counts().head(5).index
top_publisher = games['Publisher'].value_counts().head(5).index
top_platform = games['Platform'].value_counts().head(5).index
filtered_games = games[games['Genre'].isin(top_genres) & games['Year'].isin(top_years) & games['Publisher'].isin(top_publisher) & games['Platform'].isin(top_platform)]
sns.set_style('dark')
sns.countplot(x = 'Year', hue = 'Genre', data=filtered_games, order=top_years, lw=2)
```

# the number of sales of the top 5 distributors on the top 5 platforms

```python
sns.set_style('dark')
sns.countplot(x = 'Platform', hue = 'Publisher', data=filtered_games, order=top_platform, lw=2)
```

# from the top 5 platforms in the top 5 years and how many were sold

```python
sns.set_style('dark')
sns.countplot(x = 'Year', hue = 'Platform', data=filtered_games, order=top_years, lw=2)
```

# How many games of each genre were sold on each platform (Top 5)

```python
sns.set_style('dark')
sns.countplot(x = 'Platform', hue = 'Genre', data=filtered_games, order=top_platform, lw=2)
```
